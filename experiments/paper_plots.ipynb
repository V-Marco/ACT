```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Figure 4: SST and L5 FI curves

fig, ax = plt.subplots(1, 2, figsize = (9, 2))

# SST
ca = ax[0]

I = [150, 160, 210, 290, 350]
f_target = [0, 2, 32, 60, 79]
f_before = [0, 0, 0, 41, 77]
f_tuned = [0, 4, 37, 61, 78]

for f, label, color in zip(

    [f_target, f_before, f_tuned], 
    ["Target", "Before tuning", "After tuning"],
    ["blue", "gray", "red"]
    
    ):

    ca.plot(I, f, color = color, label = label)
    ca.scatter(I, f, color = color, s = 20)

ca.set_title("SST")

# L5
ca = ax[1]

I = [150, 190, 230, 270, 330]
f_target = [0, 7, 12, 15, 21]
f_before = [1, 1, 1, 1, 1]
f_tuned = [0, 9, 14, 15, 21]

for f, label, color in zip(

    [f_target, f_before, f_tuned], 
    ["Target", "Before tuning", "After tuning"],
    ["blue", "gray", "red"]
    
    ):

    ca.plot(I, f, color = color, label = label)
    ca.scatter(I, f, color = color, s = 20)

ca.set_title("L5")

for ca in ax.flatten():
    ca.set_xlabel("Current injection (pA)")
    ca.set_ylabel("Soma firing rate (Hz)")

plt.subplots_adjust(wspace = 0.4)
fig.legend(ca.get_legend_handles_labels()[0], ["Target", "Before tuning", "After tuning"], loc = "upper center", ncol = 3)
fig.suptitle(" ", y = 1.15)
```
