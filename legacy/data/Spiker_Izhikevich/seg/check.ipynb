```python
import numpy as np

def vtrap(x,y):
    if np.any(np.abs(x/y) < 1e-6):
        vtrap = y*(1 - x/y/2)
    else:
        vtrap = x/(np.exp(x/y) - 1)
    return vtrap


v = np.linspace(-150, 20, 1000)

# Activation curves from the original modfiles
#na m


#na h
alpha_nah = .07 * np.exp(-(v+65)/20)
beta_nah = 1 / (np.exp(-(v+35)/10) + 1)
        
#k n
 

nam= []
for v_ind in v:
    if v_ind < -64.5: 
        nam.append(0)
    else:
        alpha_nam = .1 * vtrap(-(v_ind+40),10)
        beta_nam =  4 * np.exp(-(v_ind+65)/18)  
        nam.append(alpha_nam / (alpha_nam + beta_nam))
        
nah = alpha_nah / (alpha_nah + beta_nah)

kdr = []
for v_ind in v:
    if v_ind < -64.5: 
        kdr.append(0)
    else:
        alpha_kn = .01*vtrap(-(v_ind+55),10) 
        beta_kn = .125*np.exp(-(v_ind+65)/80) 
        kdr.append(alpha_kn / (alpha_kn + beta_kn))
```

```python
import matplotlib.pyplot as plt

plt.plot(v, nam, label = 'nam')
plt.plot(v, nah, label = 'nah')
plt.plot(v, kdr, label = 'kdr')
plt.legend()
```
