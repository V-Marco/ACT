```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
from sklearn.cluster import KMeans
from neuron import h
from neuron_reduce.reducing_methods import (_get_subtree_biophysical_properties, find_space_const_in_cm)

def find_segment_info(key, value, seg_info_list):
    # Given a key and a value, return the corresponding seg_info
    try:
        return next((info for info in seg_info_list if info[key] == value))
    except StopIteration:
        raise ValueError(f"No segment found with {key} = {value}")


def get_elec_length(seg, frequency):
    cm, rm, ra, e_pas, q = _get_subtree_biophysical_properties(h.SectionRef(sec=seg['sec']), frequency)
    cable_space_const_in_cm = find_space_const_in_cm(seg['seg_diam']/10000, rm, ra)
    return seg['seg_L']/(cable_space_const_in_cm*10000)

def compute_electrotonic_distance(G, seg1, seg2, frequency, seg_info_list):
    # Calculate the initial distance (half of seg1's electrotonic length)
    start_dist = get_elec_length(seg1, frequency) / 2
    # Use NetworkX to get the shortest path between seg1 and seg2
    path = nx.shortest_path(G, seg1['seg_index_global'], seg2['seg_index_global'])
    # Accumulate the electrotonic lengths along the path
    for seg_index in path[1:-1]:  # Exclude the start and end segments
        seg = find_segment_info('seg_index_global', seg_index, seg_info_list)
        start_dist += get_elec_length(seg, frequency)
    # Add half of the electrotonic length of seg2 to the final distance
    end_dist = get_elec_length(seg2, frequency) / 2
    return start_dist + end_dist

def walk_graph(seg_graph, this_seg_index, terminal_nodes, analytical_electrotonic_distances, seg_to_seg_ratios, freq_hz, cell):
    # Start at the given segment
    current_seg_index = this_seg_index
    path = [this_seg_index]
    while current_seg_index not in terminal_nodes:
        # Find the next node to walk to
        next_seg_index = next(iter(seg_graph[current_seg_index]))
        path.append(next_seg_index)
        if next_seg_index not in analytical_electrotonic_distances:
            # Compute the electrotonic distance for the new segment
            seg1_info = find_segment_info('seg_index_global', current_seg_index, cell.seg_info)
            seg2_info = find_segment_info('seg_index_global', next_seg_index, cell.seg_info)
            distance = compute_electrotonic_distance(seg_graph, seg1_info, seg2_info, freq_hz, cell.seg_info)
            analytical_electrotonic_distances[next_seg_index] = distance + analytical_electrotonic_distances[current_seg_index]
        current_seg_index = next_seg_index
    return path

def compute_and_plot_electrotonic_ratios(cell, segs_to_iterate, freq_hz=0, use_kmeans=False, active=False):
    impedance = h.Impedance()
    seg_to_seg_ratios = []
    seg_graph = nx.Graph()
    analytical_electrotonic_distances = {}
    terminal_nodes = [n for n, d in seg_graph.degree() if d==1]  # nodes with degree 1 are terminal nodes

    for this_seg in segs_to_iterate:
        this_seg_ratios = []
        impedance.loc(this_seg['seg'])
        if active:
          impedance.compute(freq_hz + 1 / 9e9, 1)
        else: # passive
          impedance.compute(freq_hz + 1 / 9e9, 0)

        for other_seg in cell.seg_info:
            ratio = impedance.ratio(other_seg['seg'])
            this_seg_ratios.append((other_seg['seg_index_global'], ratio))
            seg_graph.add_edge(this_seg['seg_index_global'], other_seg['seg_index_global'], weight=ratio)

        analytical_electrotonic_distances[this_seg['seg_index_global']] = 0
        paths = walk_graph(seg_graph, this_seg['seg_index_global'], terminal_nodes, analytical_electrotonic_distances, seg_to_seg_ratios, freq_hz, cell)
        seg_to_seg_ratios.append(this_seg_ratios)

    if use_kmeans:
        kmeans = KMeans(n_clusters=2, random_state=0).fit(seg_to_seg_ratios)
        labels = kmeans.labels_
        print("KMeans Clustering labels: ", labels)

    for seg_index, distances in analytical_electrotonic_distances.items():
        ratios = seg_to_seg_ratios[seg_index]
        for path in paths:
            path_distances = [analytical_electrotonic_distances[node] for node in path]
            path_ratios = [next((r for seg_index, r in ratios if seg_index == node), None) for node in path]
            plt.plot(path_distances, path_ratios, marker='o', linestyle='-', label=f'Path {path[0]}-{path[-1]}')
        plt.title(f"Electrotonic Distances for Segment {seg_index}")
        plt.xlabel("Electrotonic Distance")
        plt.ylabel("Impedance Ratio")
        plt.legend()
        plt.show()

    return seg_to_seg_ratios, seg_graph, analytical_electrotonic_distances

# segs_to_iterate = [cell.seg_info[0]]
# seg_to_seg_ratios, seg_graph, analytical_electrotonic_distances = compute_and_plot_electrotonic_ratios(cell, segs_to_iterate)
```

```python
class KMeans:

    def __init__(self, n_clusters):
        self.n_clusters = n_clusters

        self.cluster_center_inds = None
        self.labels = None

    def fit(self, X, distance_matrix, n_iter = 100) -> None:
        # Init cluster centers randomly
        self.cluster_center_inds = np.random.choice(np.arange(len(X)), size = self.n_clusters, replace = False)

        for it in range(n_iter):
            
            # Assign points to the nearest cluster center
            self.labels = self.cluster_center_inds[np.argmin(distance_matrix[:, self.cluster_center_inds], axis = 1)]

            # Move cluster centers
            old_centers_inds = self.cluster_center_inds.copy()

            for ind_c, c in enumerate(self.cluster_center_inds):
                distance_matrix_c = distance_matrix[np.ix_(self.labels == c, self.labels == c)]
                distances_to_all_other_segments = np.sum(distance_matrix_c, axis = 1) - np.diag(distance_matrix_c)
                center_index = np.argmin(distances_to_all_other_segments)
                
                cluster_inds = np.arange(len(distance_matrix))[self.labels == c]
                self.cluster_center_inds[ind_c] = cluster_inds[center_index]

            
            # Check termination condition (it is for indices which is odd, but it's the best we can do here)
            if np.sum(np.linalg.norm(old_centers_inds - self.cluster_center_inds, 2)) < 1e-5:
                print(it)
                break
```

```python
imp_obj = h.Impedance() # create NEURON impedance hoc object
 
def calc_imp_ratio(seg1, seg2):
  '''
  computes the impedance ratio from one nrn.Segment to another
  '''
  freq_hz = 0 # choose frequency for impedance ratio computation
  imp_obj.loc(seg1) # equivalant to voltage clamping this to 1 mV
  imp_obj.compute(freq_hz + 1 / 9e9, 0) # set frequency and channel activity: 0 for passive channels, 1 for active channels
  ratio = imp_obj.ratio(seg2)

  return ratio

# Compute distance matrix
def compute_electrotonic_distance_matrix():
    all_elec_dist = np.zeros((len(all_segments), len(all_segments)))
    for i in range(len(all_segments)):
        for j in range(len(all_segments)):
            if i <= j:
                all_elec_dist[i, j] = calc_imp_ratio(all_segments[i], all_segments[j])

    distance_matrix = 1 - all_elec_dist + all_elec_dist.T - np.diag(np.diag(all_elec_dist))
    
    return distance_matrix
```

```python
# Reduce cell and store new things: can update CellModel module to do this stuff

def find_distal_sections(cell, region=str):
	'''
	Finds all terminal sections then gathers terminal apical sections that are greater than 800 microns from the soma in path length
	'''
	# find distal tuft sections:
	parent_sections=[]
	for sec in cell.all: # find non-terminal sections
		if sec.parentseg() is not None:
			if sec.parentseg().sec not in parent_sections:
				parent_sections.append(sec.parentseg().sec)
	terminal_sections=[]
	for sec in getattr(cell,region):  # check if the section is a terminal section and if it is apical tuft
		# print(h.distance(sec(0.5)))
		if region=='apic':
			if (sec not in parent_sections) and (h.distance(cell.soma[0](0.5),sec(0.5)) > 800):
				terminal_sections.append(sec)
		else:
			if (sec not in parent_sections):
				terminal_sections.append(sec)

			# print(sec, 'is a terminal section of the tuft'

	return terminal_sections

#reduce cell
if reduce_cell==True:
	#reduce complex dendritic trees to cables
	reduced_cell, synapses_list, netcons_list, txt_nr = subtree_reductor(complex_cell, synapses_list, netcons_list, reduction_frequency=0,return_seg_to_seg=True)
	print("synapses_list after NR reduction:", synapses_list)
	if expand_cable==True:
		#expand cable to idealized dendritic trees
		sections_to_expand = [reduced_cell.hoc_model.apic[0]]
		furcations_x=[0.289004]
		nbranches=[choose_branches]
		reduced_dendritic_cell, synapses_list, netcons_list, txt_ce = cable_expander(reduced_cell, sections_to_expand, furcations_x, nbranches,
																																							synapses_list, netcons_list, reduction_frequency=0,return_seg_to_seg=True)
		#remove basal dend 3d coordinates because they seem off
		for sec in reduced_dendritic_cell.dend:
			sec.pt3dclear()

		cell = CellModel(reduced_dendritic_cell,synapses_list=synapses_list,netcons_list=netcons_list,spike_threshold = 10) #NR model with apical cable converted to tree
		tufts=find_distal_sections(cell, 'apic')
		basals=find_distal_sections(cell, 'dend')
		cell._nbranch=len(tufts)

		print(cell._nbranch, "terminal tuft branches in reduced_dendritic_cell")
	else:
		#workaround since cell.all was not attribute (can update cell_model class to include this list formation)
		reduced_cell.all =[]
		for sec in [reduced_cell.soma]:
			reduced_cell.all.append(sec)
		for sec in [reduced_cell.apic]:
			reduced_cell.all.append(sec)
		for sec in reduced_cell.dend:
			reduced_cell.all.append(sec)
		for sec in reduced_cell.axon:
			reduced_cell.all.append(sec)
		#make apic sec a list (can update cell_model class to include this) (cell_model class expects cell.apic to be iterable)
		reduced_cell.apic=[reduced_cell.apic]
		#use cell_model python class
		cell=CellModel(reduced_cell,synapses_list=synapses_list,netcons_list=netcons_list,spike_threshold = 10) # neuron_reduce model
		tufts=cell.apic # Vm plot may return middle segment of trunk since it is the distal apical section.
		basals=find_distal_sections(cell, 'dend')
		cell._nbranch=len(tufts)
		print(cell._nbranch, "terminal tuft branches in reduced_cell")
else:
	basals=find_distal_sections(cell, 'dend')
	tufts=find_distal_sections(cell, 'apic')
	# cell = cell_model(complex_cell,synapses_list=synapses_list,netcons_list=netcons_list,spike_threshold = 10) # original cell
	# original cell should already be defined
	pass
```
