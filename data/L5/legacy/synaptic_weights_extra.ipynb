```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
from functools import partial
import seaborn as sns  # This is for the KDE plot
```

```python
# gmax distributions
exc_gmax_mean_0 = 0.2#0.2 # 0.45
exc_gmax_std_0 = 0.345
```

```python
exc_gmax_clip = (0,0.65)#(0, 0.7)
```

```python
exc_scalar = 0.55 # scales weight
```

```python
gmax_mean = np.log(exc_gmax_mean_0) - 0.5 * np.log((exc_gmax_std_0 / exc_gmax_mean_0) ** 2 + 1)
gmax_std = np.sqrt(np.log((exc_gmax_std_0 / exc_gmax_mean_0) ** 2 + 1))

# gmax distribution
def log_norm_dist(gmax_mean, gmax_std, gmax_scalar, size):
    val = np.random.lognormal(gmax_mean, gmax_std, size)
    s = np.clip(val, exc_gmax_clip[0], exc_gmax_clip[1])
    scaled_unclipped = gmax_scalar * np.random.lognormal(gmax_mean, gmax_std, size)
    scaled_with_std_adj = gmax_scalar * np.random.lognormal(gmax_mean, gmax_std, size)
    scaled_clipped = np.clip(val, exc_gmax_clip[0], exc_gmax_clip[1])
    return val, s, scaled_unclipped, scaled_clipped

gmax_exc_dist = partial(log_norm_dist, gmax_mean, gmax_std, exc_scalar, size = 1)
```

```python
raw, clipped, scaled_unclipped, scaled_clipped = gmax_exc_dist(size=10000)
```

```python
def plot(samples):
  # Plotting
  plt.figure(figsize=(10, 5))

  # Histogram
  plt.hist(samples, bins=50, density=True, alpha=0.6, color='b', label="Histogram")

  # KDE plot
  sns.kdeplot(samples, color='r', label="KDE")

  plt.title("Distribution Visualization")
  plt.xlabel("Value")
  plt.ylabel("Density")
  plt.legend()

  plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Compute KDE for samples
def compute_kde(samples):
    kernel = sns.kdeplot(samples).get_lines()[0]
    x = kernel.get_xdata()
    y = kernel.get_ydata()
    plt.close()  # This prevents the plot from displaying
    return x, y

# Store KDEs
kde_data = {}

kde_data["Raw"] = compute_kde(raw)
kde_data["Clipped"] = compute_kde(clipped)
kde_data["Scaled"] = compute_kde(scaled_unclipped)
kde_data["Scaled Clipped"] = compute_kde(scaled_clipped)


# Plotting stored KDEs
def plot_kdes(kde_data):
    plt.figure(figsize=(10, 5))

    for label, (x, y) in kde_data.items():
        plt.plot(x, y, label=label)

    plt.title("Multiple Distribution Visualizations")
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.legend()
    plt.show()

plot_kdes(kde_data)
```
